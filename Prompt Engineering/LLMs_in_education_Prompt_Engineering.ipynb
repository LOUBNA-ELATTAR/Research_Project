# How to write a good prompt?

Here are the key components of a good prompt:

1. A task
2. Context
3. Exemplars
4. A persona
5. Tone
6. Chain of thought

## 1. Task

The desired task should be reformulated by using an **"Action verb"**.

#### Key considerations:
- Using action verbs like generate, give, write, analyze, edit, etc.
- Clearly stating the intended outcome to guide the model effectively

<div class="alert alert-info"><h4>Examples of tasks (for both teaching and learning)</h4><p>
"Generate a creative writing lesson plan..."

"Analyze and edit the following student essay..."

"Create math word problems for a quiz on..."

"Write a reflective journal prompt for history class..."

"Generate discussion questions for a science lab on..."

"Generate a summary of the main themes in the book '1984' and provide examples to support your points..."

"Write a persuasive essay on the impact of social media on society, considering both positive and negative aspects..."
</p></div>

## 2. Context

The LLM should be provided with context by **considering the audience**, their background, and the desired success criteria. The context should be limited to what is most relevant.

#### Key considerations:
- Defining the audience precisely, specifying background details.
- Considering the audience's **environment** and what success looks like for them.

<div class="alert alert-info"><h4>Examples of context</h4><p>

"Generate a creative writing lesson plan for your grade 11 English class. The students have recently completed a unit on modern poetry, and their background includes exposure to various poetic devices and forms.."

"Create math word problems for a quiz on quadratic equations for your grade 10 algebra class. The students are familiar with factoring and solving linear equations and are now transitioning to quadratic equations.."

"As a psychology undergraduate, create a research proposal for a study on cognitive development. Draw on your coursework discussions and readings to formulate a hypothesis and design an experiment that contributes to the existing body of knowledge.."

"You're a high school student currently exploring environmental science. Write an essay reflecting on how your recent module on climate change has influenced your perspective on sustainability. Discuss personal actions you can take to contribute to a greener environment.."

</p></div>

## 3. Exemplars

Illustrating the **reasoning process** and **structure** we expect the language model to follow by including exemplars in the prompt.

#### Key considerations:
- Exemplars should demonstrate the desired reasoning and response structure
- Using specific examples that align with the task


<div class="alert alert-info"><h4>Example of providing exemplars</h4><p>

1. You are teaching linear equations to math 10 students. Develop a word problem that teenagers can relate to their own lives. The problem should include three questions (a-c). For example: "A school decides to sell t-shirts to raise money. Since there is an initial investment, if they sell 20 shirts, they will still lose \$20 overall. If they sell 100 shirts, they will make \$700."

(a) What is the slope of this relationship?

(b) What is their net profit if they do not sell any t-shirts?

(c) How many t-shirts do they need to sell to break even?



2. As a physics student, solve a projectile motion problem. For example: A ball is kicked with an initial velocity of 20 m/s at an angle of 45 degrees. Calculate the horizontal distance the ball will travel.

(a) Decompose the initial velocity into horizontal and vertical components.

(b) Use the kinematic equations to determine the time of flight.

(c) Calculate the horizontal distance traveled by the ball.

</p></div>

## 4. Persona

Defining the persona we want the LLM to embody like imagining someone we'd want instant access to for the task at hand.

#### Key considerations:
- Specifying the desired characteristics and expertise for the persona.
- Considering how the persona's traits can enhance the quality of the response.

<div class="alert alert-info"><h4>Examples of persona</h4><p>

"Act as a teacher with 20 years experience who values the unique learning styles of all students."
    
"Embody a persona that values diverse literary forms and encourages students to experiment with various poetic devices."

"Analyze and edit the following student essay as an English professor with expertise in rhetoric and composition, providing feedback that aligns with the expectations of university-level writing."

"Act as an eco-conscious biologist committed to fostering ecological awareness in your grade 12 biology class."

</p></div>


## 5. Tone

Specifying the desired tone for the response, whether it's a casual/formal voice, witty output, or enthusiasm is important.

<div class="alert alert-info"><h4>Examples of tones</h4><p>

"...Provide the feedback on the science experiment report with a formal tone. Maintain a professional demeanor, offering constructive criticism while emphasizing the importance of precision and scientific rigor."

"...Generate the discussion questions with a casual and approachable tone. Encourage students to express their opinions freely, fostering an atmosphere of open dialogue around the themes of the current novel."

"...Write the reflective essay prompt with a tone that conveys seriousness and intellectual depth. Encourage students to reflect on the historical significance of the events studied, fostering critical thinking and analytical skills."

"...Develop the creative writing assignment with a witty and playful tone. Challenge students to craft imaginative narratives, infusing humor and clever twists to make the writing process enjoyable and engaging."

</p></div>

Alternatively, we can describe the *feeling* we're going for and ask the LLM to come up with the tone:

<div class="alert alert-info"><h4>Example</h4><p>"I want to be taken seriously without coming off as too stuck up. Can you please give me a list of 5 tone keywords that I can include in a prompt for my LLM?"</p></div>

## 6. Chain of thought

By providing a process that a person might follow when accomplishing the given task. This gives the LLM a framework to reflect and improve on its response rather than constraining it to get the response right on its first try.

<div class="alert alert-info"><h4>Examples of chain of thought</h4><p>

"...Think it through step by step."

"...Follow these steps: (1) Create an outline for your paragraph, (2) Edit your outline, (3) Write the paragraph."

</p></div>

## Complete example

<div class="alert alert-info"><h4>Creating an engaging math lesson plan</h4><p>

You are a math teacher who excels at making real-world connections to math. Design a comprehensive lesson plan for an 80 minutes grade 8 math class that introduces the concept of linear equations through graphs and tables of values. Incorporate real-world examples to make the lesson engaging and relevant. Including the following sections in the lesson plan: (1) the learning objectives (2) step-by-step activities, and (3) the assessment methods. Additionally, highlight any interactive elements or visual aids that will enhance student understanding.

</p></div>

**ChatGPT Answer:**
**Grade 8 Math Lesson Plan: Introducing Linear Equations through Graphs and Tables**

### Learning Objectives:

1. **Conceptual Understanding:**
   - Students will comprehend the fundamental principles of linear equations.
   - Students will grasp the relationship between graphs, tables of values, and linear equations.

2. **Graphical Interpretation:**
   - Students will learn to interpret and draw linear graphs.
   - Students will connect real-world scenarios to linear equations and their graphical representations.

3. **Analytical Skills:**
   - Students will be able to create tables of values for linear equations.
   - Students will develop analytical skills to identify slope and y-intercept from graphs.

### Step-by-Step Activities:

**Introduction (10 minutes):**
   - Begin with a real-world scenario: "Imagine a scenario where you earn $10 per hour for a part-time job. Let's explore how your earnings can be represented using linear equations."
   - Introduce the concept of linear equations verbally and through a brief visual presentation.

**Activity 1 - Exploring Graphs (20 minutes):**
   - Distribute graph paper and ask students to plot the linear equation \(y = 2x + 5\).
   - Discuss the meaning of slope and y-intercept in the context of the graph.
   - Use an interactive whiteboard to draw graphs collaboratively, emphasizing key points.

**Activity 2 - Creating Tables of Values (20 minutes):**
   - Provide scenarios (e.g., distance vs. time) and guide students in creating tables of values.
   - Discuss how the table of values relates to the linear equation.
   - Utilize a digital tool or interactive software to input equations and generate corresponding tables dynamically.

**Activity 3 - Real-World Connections (20 minutes):**
   - Introduce real-world examples such as calculating expenses for a phone plan or predicting future savings.
   - Have students create their linear equations based on these scenarios and discuss their findings.
   - Encourage peer collaboration and discussion.

**Conclusion and Summary (10 minutes):**
   - Summarize key concepts and emphasize the practical applications of linear equations.
   - Assign a small homework task that involves finding linear equations in real-life scenarios.

### Assessment Methods:

1. **Formative Assessment:**
   - Monitor student participation during activities and discussions.
   - Review students' graphs and tables for accuracy and understanding.

2. **Individual and Group Tasks:**
   - Evaluate the tables of values created by students during Activity 2.
   - Assess the real-world connections made by students in Activity 3.

3. **Exit Ticket:**
   - Distribute an exit ticket with a real-world scenario, asking students to create a corresponding linear equation and graph.
   - Use the exit ticket to gauge individual understanding before the next class.

### Visual Aids and Interactive Elements:

1. **Interactive Whiteboard:**
   - Use the whiteboard for collaborative graph drawing.
   - Display real-world scenarios and equations dynamically.

2. **Graphing Software:**
   - Utilize graphing software to illustrate linear equations and generate tables of values.
   - Allow students to experiment with different equations.

3. **Real-World Scenarios:**
   - Incorporate visual aids such as images or charts representing real-world scenarios.
   - Use props or visuals to enhance engagement during the introduction and real-world connections.

By incorporating real-world scenarios, interactive elements, and clear learning objectives, this lesson plan aims to make the concept of linear equations tangible and relevant for grade 8 students.

# Back-and-forth interactions with an LLM


One of the most powerful ways to interact with an LLM is by providing feedback through a back-and-forth conversation. This will allow the user to provide input for the LLM to use to either improve its response or dive deeper into specific aspects of its responses.

With GPT4 for example, we can even provide an initial prompt with instructions for ChatGPT to facilitate a back-and-forth conversation.

For instance, the following prompt is a useful way to develop effective explanations of concepts that the students may find tricky to understand:

<div class="alert alert-info"><h4>Concept explanation, examples and analogies</h4><p>

"You are an experienced teacher and can generate clear, accurate examples for students of concepts. I want you to ask me two questions.
The first question is, "What concept do you want explained?" Wait for me to answer before asking me the second question: "Who is the audience for the explanation?"
Then look up the concept and examples of the concept. Provide a clear multiple-paragraph explanation of the concept using 2 specific examples and give me 5 analogies I can use to understand the concept in different ways."
    
</p></div>

The below prompt can help you use ChatGPT to develop a syllabus for your class, which will likely require several iterations of back-and-forth conversation.
